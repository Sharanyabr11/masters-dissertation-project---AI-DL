```python
from ultralytics import YOLO

model = YOLO("yolov8m.pt")  # Load a pre-trained model
```

```python
model_path = '/content/runs/detect/train2/weights/best.pt'  # Path to your trained model
dataset_path = '/content/YOLO-chemical-structures-12/data.yaml'  # Path to your data.yaml file

# Load the trained YOLO model
model = YOLO(model_path)

# Perform validation using the specified dataset
results = model.val(data=dataset_path)

# Print validation results
print(results)
```

```python
test_images_path = '/content/YOLO-chemical-structures-12/test/images'
# Run the prediction
results = model.predict(source=test_images_path, conf=0.25)

# Display the results
for result in results:
    result.show()
```
